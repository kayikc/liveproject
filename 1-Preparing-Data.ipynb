```python
import pandas as pd
```

```python
df_temp = pd.read_csv("Milestone 1 Dataset\\hr_temp_20170201-20200131_subset.csv" )
df_energy = pd.read_csv("Milestone 1 Dataset\\hrl_load_metered - 20170201-20200131.csv")
```

```python
df_energy.head()
```

```python
df_temp.head()
```

```python
cols_wanted = {"date": df_temp['DATE'], "mw": df_energy['mw'], "temp": df_temp["HourlyDryBulbTemperature"]}
```

```python
df = pd.DataFrame(cols_wanted)
df.date = df.date.astype('datetime64[ns]')
df.dtypes
```

```python
df
```

```python
df['hour'] = df['date'].dt.hour
df['weekday'] = df['date'].dt.dayofweek
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
df
```

```python
df.set_index('date')
```

```python
df['temp'] = df['temp'].interpolate(method='linear')
```

```python
mask = df['date'] < '2020-01-01'
df_training = df[mask]
df_training.head(-5)
```

```python
df_testing = df[~mask]
df_testing.head(5)
```

```python
df_training.to_csv('df_training')
df_testing.to_csv('df_testing')
```
